## Standardize ratings of comics 

source: https://en.wikipedia.org/wiki/Marvel_Comics_rating_system

```python
df["Rating"] = df["Rating"].replace(' T', ' Rated T')
df["Rating"] = df["Rating"].replace(' Parental Guidance', ' Rated T')
df["Rating"] = df["Rating"].replace(' T+', ' Rated T+')
df["Rating"] = df["Rating"].replace(' Marvel Psr', ' Rated T+')
df["Rating"] = df["Rating"].replace(' Ages 10 & Up', ' Rated T')
df["Rating"] = df["Rating"].replace(' Ages 9+', ' Rated T')
df["Rating"] = df["Rating"].replace(' None', ' No Rating')
df["Rating"] = df["Rating"].replace('None', ' No Rating')
df["Rating"] = df["Rating"].replace(' Not In Oracle', ' No Rating')
df["Rating"] = df["Rating"].replace(' Parental Advisory/Explicit Content', ' Parental Advisory')
df["Rating"] = df["Rating"].replace(' Parental Supervision', ' Parental Advisory')
df["Rating"] = df["Rating"].replace(' Parental Advisoryslc', ' Parental Advisory')
df["Rating"] = df["Rating"].replace(' Parental Guidance', ' Parental Advisory')
df["Rating"] = df["Rating"].replace(' Marvel Psr+', ' Parental Advisory')
df["Rating"] = df["Rating"].replace(' Explicit Content', ' Explicit Content')
df["Rating"] = df["Rating"].replace(' Mature', ' Explicit Content')
df["Rating"] = df["Rating"].replace(' Max', ' Explicit Content')
df["Rating"] = df["Rating"].replace(' A', ' Explicit Content')
df["Rating"] = df["Rating"].replace(' Rated A', ' Explicit Content')
```

```python
df['Rating'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.countplot(x="Rating",data=df)
plt.xticks(rotation=45)
plt.show()
```

## Which year had the most comics released past the year 2000s?

```python
df.head(2)
```

```python
df[['Month', 'Year']] = df['publish_date'].str.split(', ', expand = True)
df
```

```python
df['Year'] = pd.to_numeric(df['Year'])
df.dtypes
```

```python
df['Year'].value_counts()
```

```python
cent_20s = df[df['Year'] >= 2000]

sns.countplot(x="Year",data = cent_20s)
plt.xticks(rotation=90)
plt.show()
```

## Which year had the most expensive comic past the year 2000s?

```python
df['Price'].value_counts()
```

```python
df["Price"] = df["Price"].replace('None', ' Free')
df['Price'].value_counts()
```

```python
df["Price"] = df["Price"].str.replace('$', '')
df["Price"] = df["Price"].str.replace('Free', '0.00')
df["Price"] = pd.to_numeric(df['Price'])
df
```

```python
df["Price"].max()
```

```python
df[df["Price"] == 99.99]
```

```python
cent_20s["Price"] = cent_20s["Price"].str.replace('$', '')
cent_20s["Price"] = cent_20s["Price"].replace('None', ' Free')
cent_20s["Price"] = cent_20s["Price"].str.replace('Free', '0.00')
cent_20s["Price"] = cent_20s["Price"].astype(float)

cent_20s[['Year','Price']].groupby('Year').Price.max()
```

```python
plt.bar('Year', 'Price', data = cent_20s)
plt.show()
```

The year 2008 had the most expensive comic with a price of $99.99 which was the series of "Howard the Duck"
